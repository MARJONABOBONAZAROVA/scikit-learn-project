```python
import numpy as np #all libraries
import pandas as pd
import matplotlib.pyplot as plt

#scikit-learn model
from sklearn.linear_model import LinearRegression,Ridge,Lasso, LogisticRegression
from sklearn.ensemble import RandomForestRegressor,RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier

#scikit-learn utilities
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_regression, make_classification

#for ignore  warning
import warnings
warnings.filterwarnings("ignore")

%matplotlib inline
print("all libraries running")
```

```python
#regression uchun dataset yaratish
np.random.seed(42)
np_samples = 500

#features
size = np.random.randint(500, 3500, np_samples) #square feet
bedrooms = np.random.randint(1, 6, np_samples)#number of bedrooms
age = np.random.randint(0,30, np_samples)# age of hause
distance_to_city = np.random.randint(1, 20, np_samples)#distance to city

#target
price = (size * 200) + (bedrooms * 15000) - (age * 2000) - (distance_to_city * 500) + np.random.normal(0, 3000, np_samples)

#create dataframe
house_data = pd.DataFrame({
    "size": size,
    "bedrooms": bedrooms,
    "age": age,
    "distance_to_city": distance_to_city,
    "price": price

})

print("house price dataset")
print(house_data.head())
print(f"\ndataset shape:{house_data.shape}")
```

```python
#prepare data for modeling
X = house_data.drop("price", axis=1)
Y = house_data["price"]

#split data 
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
print(f"Traine samples: {len(X_train)}")
print(f"Test samples: {len(X_test)}")
```
